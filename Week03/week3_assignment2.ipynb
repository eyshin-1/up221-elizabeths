## Mapping Data

```python
import geopandas as gpd
```

**Import geojson file from LA Times for Census Tracts (2020)**

```python
tracts=gpd.read_file('Data/Census_Tracts_2020.geojson')
tracts.head()
```

```python
tracts.plot(figsize=(12,10))
```

```python
tracts.info(verbose=True, show_counts=True)
```

```python
tracts = tracts[['CT20','geometry']]
tracts.head()
```

```python
tracts.loc[:, 'FIPS'] = '06' + '037' + tracts['CT20']
```

```python
tracts.head()
```

## Merge Census Tract Data and Race Data
**Create new data set that merges df from ACS and geojson from LA Times**

```python
tracts_race=tracts.merge(df2,on="FIPS")
```

```python
tracts_race.head()
```

```python
tracts_race.plot(figsize=(12,10),
                 column='Asian Alone',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
tracts_race.plot(figsize=(12,10),
                 column='Black or African American Alone',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
tracts_race.plot(figsize=(12,10),
                 column='American Indian and Alaska Native Alone',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
tracts_race.plot(figsize=(12,10),
                 column='Native Hawaiian and Other Pacific Islander Alone',
                 legend=True, 
                 scheme='NaturalBreaks')
```
