The **Los Angeles County Vision Zero Collision Concentration Corridors (CCC)** data refers to streets or corridors in Los Angeles with a high concentration of traffic collisions. The Vision Zero initiative sets a goal of zero traffic-related fatalities and injuries by 2035 through infrastructure changes, policies, and enforcement. In this assignment, I am plotting these corridors over a base map of Los Angeles to reveal streets of interest for safety-related interventions.

List of desired column names

Subset based on desired columns

```python
desired_columns = ['STREET','DPWSCORE_M','RANKING','LIMITS','geometry']

visionzero[desired_columns].sample(5)
```

Declare dataframe new variable = visionzero_trimmed

```python
visionzero_trimmed = visionzero[desired_columns].copy()
visionzero_trimmed
```

Filter data by street

```python
visionzero_trimmed[visionzero_trimmed.STREET == 'Whittier Bl']
```

Plot new filtered data

```python
visionzero_trimmed.plot()
```

Add contexily for base map

```python
import contextily as ctx
```

Ensure projections are the same

```python
visionzero_trimmed = visionzero_trimmed.to_crs(epsg=3857)
```

Create plot

```python
ax = visionzero_trimmed.plot(  # create axes
    figsize=(20, 12),    # size of plot
    column='STREET',     # column that defines the color of the dots
    legend=True,         # add a legend
    legend_kwds={
        'loc': 'upper right',
        'bbox_to_anchor': (1, 1)  # position the legend
    }
)

# add basemap
ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron)
```
